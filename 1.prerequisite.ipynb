```python
from mysql.connector import Error
from mysql.connector.pooling import MySQLConnectionPool

try:
  dbconfig={"database":"little_lemon_db", "user":"root", "password":"123123"}

  pool_a = MySQLConnectionPool(pool_name = "ll_pool_a",
                              pool_size = 2,
                            **dbconfig)

  # Get the connection from the connection pool "pool"
  print("Getting a connection from the pool.")
  connection = pool_a.get_connection()
  
  # Creating cursor object
  cursor = connection.cursor()
  
  # Create the stored procedure query as a Python string
  peak_hours_stored_procedure_query = """
  CREATE PROCEDURE IF NOT EXISTS PeakHours ()
  BEGIN
  SELECT HOUR(BookingSlot), COUNT(*)
  FROM Bookings 
  GROUP BY HOUR(BookingSlot)
  ORDER BY COUNT(*) DESC; 
  END
  """
  
  cursor.execute(peak_hours_stored_procedure_query)
  
  cursor.callproc("PeakHours")
  
  results = next(cursor.stored_results())
  
  dataset = results.fetchall()
  
  for data in dataset:
    print(data)
  
  guest_status_stored_procedure_query = """
  CREATE PROCEDURE IF NOT EXISTS GuestStatus ()
  BEGIN
  SELECT CONCAT(b.GuestFirstName," ", b.GuestLastName) AS GuestName, 
  CASE
    WHEN e.Role IN ("Manager", "Assistant Manager") THEN "Ready to pay"
    WHEN e.Role = "Head Chef" THEN "Ready to serve"
    WHEN e.Role = "Assistant Chef"  THEN "Preparing Order"
    WHEN e.Role = "Head Waiter" THEN "Order served"
    ELSE "Unhandled"
  END AS Status
  FROM Employees e 
  LEFT JOIN Bookings b
  ON e.EmployeeID = b.EmployeeID;
  END
  """
  
  cursor.execute(guest_status_stored_procedure_query)
  
  cursor.callproc("GuestStatus")
  
  results = next(cursor.stored_results())
  
  dataset = results.fetchall()
  
  for data in dataset:
    print(data)
    
  # Put the connection back to the pool    
  print("\nReturning the connection back to the pool.")
  connection.close()
  
  print("The connection is placed back into the pool for the next user to connect.")
except Error as e:
    print(f"Error: {e}")
```
